# Diabetes Prediction - Early Stage

* We have data about both males and females
* Data is symptoms for Diabetes and thier acutual results
* Reseach paper for the dataset used - https://ieeexplore.ieee.org/document/9856420
* This model may help in Medical domain

### Data dictionary

1. `Age`: Age of the patient in years.
Gender: Gender of the patient. Typically 'Male' or 'Female'.
2. `Polyuria`: Indicates whether the patient experiences excessive urination. Yes if present, No if not.
3. `Polydipsia`: Indicates whether the patient experiences excessive thirst. Yes if present, No if not.
4. `Sudden weight loss`: Indicates whether the patient has experienced sudden weight loss. Yes if present, No if not.
5. `Weakness`: Indicates whether the patient experiences weakness. Yes if present, No if not.
6. `Polyphagia`: Indicates whether the patient experiences excessive hunger. Yes if present, No if not.
7. `Genital thrush`: Indicates whether the patient has genital thrush. Yes if present, No if not.
8. `Visual blurring`: Indicates whether the patient experiences visual.
9. `blurring`. Yes if present, No if not.        
10. `Itching`: Indicates whether the patient experiences itching. Yes if present, No if not.
11. `Irritability`: Indicates whether the patient experiences irritability. Yes if present, No if not.
12. `Delayed healing`: Indicates whether the patient experiences delayed healing of wounds. Yes if present, No if not.
13. `Partial paresis`: Indicates whether the patient experiences partial paresis (muscle weakness). Yes if present, No if not.
14. `Muscle stiffness`: Indicates whether the patient experiences muscle stiffness. Yes if present, No if not.
15. `Alopecia`: Indicates whether the patient experiences alopecia (hair loss). Yes if present, No if not.
16. `Obesity`: Indicates whether the patient is obese. Yes if present, No if not.
17. `Class`: Classification of the patient, typically 'Positive' for diabetes or 'Negative' for non-diabetes.

## Splittig the data for training and testing

```python
# importing train test split
from sklearn.model_selection import train_test_split

# set random seed
np.random.seed(42)

# Seperating Target and Features
x_early = diabetes_early_df.drop('Class', axis = 1)
y_early = diabetes_early_df['Class']

# Splitting data into train and test
x_train, x_test, y_train, y_test = train_test_split(x_early, y_early, test_size = 0.2, random_state = 42)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### Models Used :
  *  Logistic Regression (LR)
  * Naive Bayes (NB)
  * Support Vector Machine (SVM)
  * Multi-Layer Perceptron (MLP)
  * Random Forest Classifier (RF)
  * XGBoost

```python
# Importing models
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
```

```python
# Make model dictionary
models = {'Logistic Regression' : LogisticRegression(),
          'Naive Bayes' : GaussianNB(),
          'Support Vector Machine' : SVC(),
          'Multi-Layer Perceptron' : MLPClassifier(),
          'Random Forest' : RandomForestClassifier(),
          'XG-Boost' : xgb.XGBClassifier()}

# Create a function to fit and score models :

def fit_and_score(models, x_train, x_test, y_train, y_test):
    """"
    This function fits and evaluates the models
    models : dictionary of models
    x_train : training data (no labels)
    x_test : testing data (no labels)
    y_train : trainign data (labels)
    y_test : testing data (labels)
    """
    # random seed
    np.random.seed(42)
    # Make score dictionary
    model_scores = {}
    # Loop through models
    for name, model in models.items():
        # Fit the model
        model.fit(x_train, y_train)
        # Append the scores in dictionary
        model_scores[name] = model.score(x_test,y_test)
    return model_scores
```

```python
# Get cross-val-score function
from sklearn.model_selection import cross_val_score

def get_cross_val_score(models, x, y):
  """"
  This function fits and evaluates the models based cross val score
  models : dictionary of models
  x : data with no labels
  y : label data
  """
  # random seed
  np.random.seed(42)
  # Make score dictionary
  cross_scores = {}
  # Loop through models
  for name, model in models.items():
      cv_acc =  cross_val_score(model,
                        x,
                        y,
                        cv = 5,
                        scoring = "accuracy")
      cross_scores[name] = cv_acc.mean()
  return cross_scores
```

```python
# Get scores
model_scores = fit_and_score(models = models,
                             x_train = x_train,
                             x_test = x_test,
                             y_train = y_train,
                             y_test = y_test)

model_scores
```

```python
# Get cross val scores
cross_scores = get_cross_val_score(models = models,
                                   x = x_early,
                                   y = y_early)
cross_scores
```

```python
# Model comparison visualization
model_compare = pd.DataFrame(model_scores, index = ['Accuracy'])
model_compare.T.plot.bar(figsize = (12,6));
plt.xticks(rotation = 0);
```

```python
# Visualize Cross-val score
cross_compare = pd.DataFrame(cross_scores, index = ['Cross Val Score'])
cross_compare.T.plot.bar(figsize = (12,6));
plt.xticks(rotation = 0);
```

## Pre-Hyperparameter Tunning Results

* $Random Forest Classifier$ has the highest `accuarcy` among all the models - **0.98**
* $Random Forest Classifier$ has the highest `cross-val-score` among all models - **0.975**
* Hence, we will try to improve this model with `hyperparameter tuning`

```python
from sklearn.model_selection import GridSearchCV
```

```python
# create hyperparameters for RFC
rf_grid = {'n_estimators' : np.arange(10,500,50),
           'max_depth' : [None, 3,5,10],
           'min_samples_split' : np.arange(2,10,2),
           'min_samples_leaf' : np.arange(1,10,2)}
```

```python
rs_rf = GridSearchCV(RandomForestClassifier(),
                           param_grid = rf_grid,
                           cv = 3,
                           verbose = True)
```

```python
rs_rf.fit(x_train, y_train)
```

```python
rs_rf.best_params_
```

```python
# getting best params
clf = RandomForestClassifier(max_depth = 10,
 min_samples_leaf =  1,
 min_samples_split =  2,
 n_estimators =  110)
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
y_preds = clf.predict(x_test)
```
